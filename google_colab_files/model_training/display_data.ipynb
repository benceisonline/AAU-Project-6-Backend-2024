This notebook is for displaying AUC score data for the LightFM models

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Define paths to folders
folders = [
    "/content/drive/MyDrive/SW6/Colab notebooks/LightFM/Sprint 5 Research/Exported Data/latent-factors-5-small/",
    "/content/drive/MyDrive/SW6/Colab notebooks/LightFM/Sprint 5 Research/Exported Data/latent-factors-10-small/",
    "/content/drive/MyDrive/SW6/Colab notebooks/LightFM/Sprint 5 Research/Exported Data/latent-factors-20-small/",
    "/content/drive/MyDrive/SW6/Colab notebooks/LightFM/Sprint 5 Research/Exported Data/5-latent-factors/4000-epochs",
    "/content/drive/MyDrive/SW6/Colab notebooks/LightFM/Sprint 5 Research/Exported Data/10-latent-factors/4000-epochs",
    "/content/drive/MyDrive/SW6/Colab notebooks/LightFM/Sprint 5 Research/Exported Data/20-latent-factors/4000-epochs",]

# Loop through each folder
for folder_path in folders:
    # Load README.md content
    readme_path = os.path.join(folder_path, "README.md")
    with open(readme_path, "r") as readme_file:
        readme_content = readme_file.read()
    print("README.md contents for folder:", folder_path)
    print(readme_content)

    # Load auc_scores_per_epoch.csv
    csv_path = os.path.join(folder_path, "auc_scores_per_epoch.csv")
    df = pd.read_csv(csv_path)

    # Plot data
    plt.figure(figsize=(10, 5))
    plt.plot(df["Epoch"], df["Train AUC"], label="Train AUC")
    plt.plot(df["Epoch"], df["Test AUC"], label="Test AUC")
    plt.xlabel("Epoch")
    plt.ylabel("AUC Score")
    plt.title("AUC Scores per Epoch")
    plt.legend()
    plt.grid(True)
    plt.show()
```
